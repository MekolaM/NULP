# Домашнее задание к лекции "Функции"

Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def doc_id(id):
    for document in documents:
        if document['number'] == id:
            return document['name']
    return 'Не знайдено'
def doc_shell(id):
    for directory in directories.items():
        if id in directory[1]:
            return directory[0]
    return 'Не знайдено'
def doc_print():
    for document in documents:
        print(f"тип: {document['type']}, номер: {document['number']}, имя: {document['name']}, \
полка: {doc_shell(document['number'])}")
def add_shell(id):
    if id not in directories.keys():
        directories[id]=[]
        print("Додано")
    else:
        print("Така полка вже є, список полок:",directories.keys())
def del_shell(id):
    if id in directories.keys():
        if directories[id] == []:
            del directories[id]
            print('Видалено')
        else:
            print('Полка не пуста')
    else:
        print('Полка не знайдена, список полок: ',directories.keys())
def add_doc(type,id,name,shell):
    if shell in directories.keys():
        documents.append({'type':type,'number':id,'name':name})
        directories[shell].append(id)
        print('Додано')
        doc_print()
    else:
        print('Полка не знайдена, список полок: ',directories.keys())
def del_doc(id):
    for i in range(len(documents)):
        if documents[i]['number'] == id:
            del documents[i]
            print('Видалено')
            doc_print()
            return 1
    print('id не знайдено')
    doc_print()
def move_doc(id,shell):
    if shell in directories.keys():
        for key,value in directories.items():
            if id in value:
                directories[key].remove(id)
                directories[shell].append(id)
                print('Змінено')
                doc_print()
                return 1
        else:
            print('Документ не знайдено')
    else:
        print('Полка не знайдена, список полок: ',directories.keys())
def main():
    while True:
        user_input=input('Введіт команду')
        if user_input == 'p':
            print(doc_id(input('Введіт ід')))
        elif user_input == 's':
            print('Документ на полці',doc_shell(input('Введіт ід')))
        elif user_input == 'l':
            doc_print()
        elif user_input == 'ads':
            add_shell(input('Введіт ід'))
        elif user_input == 'ds':
            del_shell(input('Введіт ід'))
        elif user_input == 'ad':
            add_doc(input('Введіт тип'),input('Введіт ід'),input('Введіт імя'),input('Введіт полку'))
        elif user_input == 'd':
            del_doc(input('Введіт ід'))
        elif user_input == 'm':
            move_doc(input('Введіт ід'),input('Введіт полку'))
        else:
            break
main()
````
